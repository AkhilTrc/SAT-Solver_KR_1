```python
import math
import re

sudoku_rules_path = "sudo_rules"  
sudoku_data_path = "sudo_data"
output_path = "convert_data"
sudoku_rules = {
    4 : "sudoku-rules-4x4.txt",
    9 : "sudoku-rules-9x9.txt",
   16 : "sudoku-rules-16x16.txt"
}

# 10-16 to A-G
def letter_gen(x):
    if x >= 10:
        return chr(ord('A') + x - 10)
    else:
        return str(x)

# A-G to 10-16
def number_gen(tok):
    if tok in {'A','B','C','D','E','F','G'}:
        return int(ord(tok)-ord('A') + 10)
    return int(tok)
 
def get_dimacs_string(line):
    sudoku_string = ""
    cnt = 0
    sudoku_size =  math.isqrt(len(line))   
    for tok in line:
        cnt += 1
        if tok.isalnum():
            r = (cnt - 1) // sudoku_size + 1
            c = cnt % sudoku_size if cnt % sudoku_size != 0 else sudoku_size
            v = number_gen(tok)
            pseudo_base = (sudoku_size + 1) if sudoku_size > 10 else 10
            sudoku_string += "{} 0\n".format(r * pseudo_base**2 + c * pseudo_base + v)
    return sudoku_string

def get_dimacs_strings_from_file(n, puzzles_file):  
    f = open(sudoku_data_path+"/"+puzzles_file, "r")
    f.seek(0)
    lines = f.readlines()  
    dimacs = []  
    k = 0
    for line in lines:
        if k < n:
            dimacs.append(get_dimacs_string(line)) 
            k = k + 1

    puzzle_size = math.isqrt(len(lines[0]))
    f = open(sudoku_rules_path+"/"+sudoku_rules[puzzle_size])  
    rules = f.read()
    f.close()
    return dimacs, rules, puzzle_size
```
